```python
import pandas as pd
```

```python
df = pd.read_csv('Emissions_Agriculture_Enteric_Fermentation_E_All_Data_(Normalized).csv', encoding = "ISO-8859-1")
```

```python
pd.unique(df['Unit'])
```

```python
df = df[df['Unit'].str.contains('kilotonnes') == False]
```

```python
pd.unique(df['Unit'])
```

```python
pd.unique(df['Item'])
```

```python
pd.unique(df['Source Code'])
```

```python
# Drop unneeded columns
list(df.columns)
```

```python
drop_cols = ['Area Code','Area Code (M49)', 'Element Code', 'Year Code', 'Source Code','Unit']
```

```python
df = df.drop(columns=drop_cols)
```

```python
pd.unique(df['Area'])
```

```python
# Since we just want the categorical info we can drop value 
# df = df.drop(columns='Value')
```

```python
df
```

```python
# Get date ranges for each item, depending on source 
source_years = pd.DataFrame(df.groupby('Source')['Year'].unique().apply(list))
source_years = source_years.reset_index()
source_years

# Output 
source_years.to_csv('EF_source_year.csv', index=False)
```

```python
source_years
```

```python
item_area_years = pd.DataFrame(df.groupby(['Item', 'Area','Source'])['Year'].apply(list))
item_area_years = item_area_years.reset_index()
item_area_years

# Output 
item_area_years.to_csv('EF_Item_Area_Source_Year.csv', index = False)
```

```python
source_item_year = pd.DataFrame(df.groupby(['Source','Item'])['Year'].unique().apply(list))
source_item_year = source_item_year.reset_index()
source_item_year.to_csv('EF_Source_Item_Year.csv', index=False)
```

```python
source_country_year = pd.DataFrame(df.groupby(['Source','Area'])['Year'].apply(list))
source_country_year = source_country_year.reset_index()
source_country_year

source_country_year.to_csv('EF_Source_Country_Year.csv', index = False)
```
